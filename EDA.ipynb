```python
import re # regular expression
# https://pymotw.com/2/re/
import nltk
import string
from nltk.corpus import stopwords

stop = set(stopwords.words('english')) # set of stopwords in english
sno = nltk.stem.SnowballStemmer('english') # initializaing snowball stemmer

# this fn will replace an html tag with space
def cleanthtml(sentence): # fn to clean the word of any html tags
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ',sentence) # substitute
    return cleantext

def cleanpunc(sentence):
    cleaned = re.sub(r'[?|!|\'|"|#]', r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/)]',r' ',cleaned)
    return cleaned
```

```python
def clean_tweet(data):
    clean_list = []
    for tweet in data.values:
        tweet = cleanthtml(tweet)
        tweet = cleanpunc(tweet)
        sent = []
        for word in tweet.split(" "):
            if((word.isalpha()) & (len(word)>2)):
                if(word.lower() not in stop):
                    word = (sno.stem(word.lower()))
                    sent.append(word)
        clean_list.append(" ".join(sent))
    return clean_list
```

```python
train_clean = clean_tweet(train_df.tweet)
```

```python
train_df['clean_tweet'] = np.array(train_clean)
```

```python
test_clean = clean_tweet(test_df.tweet)
test_df['clean_tweet'] = np.array(test_clean)
```

### Feature Encoding
- we will try tfidf

```python
# splitting data
from sklearn.model_selection import train_test_split
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
tweet_vect = TfidfVectorizer(ngram_range=(1,2))
tweet_vect.fit(train_df.clean_tweet)
train_tfidf = tweet_vect.transform(train_df.clean_tweet)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_tfidf, train_df.label, test_size = 0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### Machine learning models

```python
from sklearn.decomposition import TruncatedSVD
svd_train = TruncatedSVD(n_components=2500)
train_svd  = svd_train.fit_transform(train_tfidf)
```

```python
print("svd_train shape : ",train_svd.shape)
X_train, X_test, y_train, y_test = train_test_split(train_svd, train_df.label, test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegressionCV
```

```python
alpha = [0.001,0.01,0.1,0.5,1,5,10,100]
nb  = LogisticRegressionCV(Cs=alpha, class_weight="balanced", cv=3, penalty="l2",n_jobs=-1)
nb.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
from sklearn.metrics import f1_score
score = f1_score(y_test, y_pred)
print(score)
```
